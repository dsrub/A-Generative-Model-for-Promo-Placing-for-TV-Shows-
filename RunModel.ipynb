```python
from DMAGenerativeModel import DMAGenerativeModel, SortRatios
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
### fetch data, filter and fit model

model = DMAGenerativeModel(['SPONGEBOB SQUAREPANTS'])
model.FetchData()
model.FilterData()
model.MLfitting(15)
```

```python
### plot

### compute ratios with given Delta and sort
ratios = model.ComputeRatios(.0005)
expt, err, dmas = SortRatios(ratios)

expt = expt[:45]
err = err[:45]
dmas = dmas[:45]


fig, ax = plt.subplots(figsize=(40, 10))

x_pos = np.arange(len(dmas))

ax.bar(x_pos, expt, yerr=err, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_xticks(x_pos)
ax.set_xlim(-1, len(dmas))
ax.set_ylabel('Expected lift in WMV (%)', size = 30)
ax.set_title('Spongebob', size = 30)
ax.tick_params(axis = 'y', which = 'major', labelsize = 25)
ax.set_xticklabels(dmas, rotation=90, size=19);
fig.savefig('spongebob_lift.pdf', bbox_inches='tight');
```
